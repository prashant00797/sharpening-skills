# 🧩 Q1. User Profile Endpoint with Conditional Response

## 📌 Scenario

You are building a **user profile API** for a healthcare app.

Create a `GET` endpoint:

```
/users/{user_id}
```

### Requirements:

* Accept:

  * `user_id` as a **path parameter**
  * `include_health_data` as an **optional query parameter** (default: `False`)
* Use **mock data** (Python dictionary), no database

---

## 🎯 Expected Behavior

* If `include_health_data = False`
  → Return **basic user info**

* If `include_health_data = True`
  → Return **full user info including health data**

---

## 🧱 Response Models

### 🟢 UserBasic

* `name`
* `email`

---

### 🔴 UserFull

* `name`
* `email`
* `blood_group`
* `allergies`

---

## 🧠 Concepts Tested

* Path Parameters
* Query Parameters
* Pydantic Response Models
* Conditional Response Handling

---

## 💡 Hint

* Define **two separate Pydantic models**
* Use an `if/else` inside the endpoint:

  * return basic model OR full model
* Path param → part of URL
* Query param → function argument with default value
